```python
import matplotlib.pylab as plt 
%matplotlib inline
import numpy as np
import rubin_sim.maf_proto as maf
import pandas as pd
import sqlite3
import healpy as hp
import copy
import datetime

from rubin_sim.data import get_baseline
from os.path import basename
from rubin_scheduler.scheduler.utils import get_current_footprint
```

```python
# Read in the current baseline
db_file = get_baseline()
run_name = basename(db_file).replace('.db', '')
con = sqlite3.connect(db_file)
df = pd.read_sql('select * from observations;', con)
con.close()
# Convert to a numpy array
visits_array = df.to_records(index=False)
```

```python
nside = 64
fp, lab = get_current_footprint(nside=nside)
wfd_indx = np.where(lab == "lowdust")[0]
# Quick Coadded depth in r-band
sub_data = visits_array[np.where((visits_array["band"] == "r") &
                                 (visits_array["night"] < 61))]
sl = maf.Slicer(nside=nside, missing=-26)
metric = maf.CoaddM5Metric("r")
array = sl(sub_data, metric, indx=wfd_indx)
array
```
