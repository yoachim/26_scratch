```python
import pandas as pd
import numpy as np
```

```python
orbits = pd.read_csv("/Users/yoachim/rubin_sim_data/orbits/granvik_5k.txt", delimiter=" ")
```

```python
orbits
```

```python
out_orbits = pd.DataFrame()
```

```python
out_orbits["ObjID"] = np.arange(orbits["objId"].size)
out_orbits["a"] = orbits["a"]
out_orbits["e"] = orbits["e"]
out_orbits["inc"] = orbits["inc"]
out_orbits["node"] = orbits["Omega"]
out_orbits["argPeri"] = orbits["argPeri"]
out_orbits["ma"] = orbits["meanAnomaly"]
out_orbits["epochMJD_TDB"] = orbits["epoch"]
out_orbits["FORMAT"] = "KEP"
```

```python
out_orbits
```

```python
out_orbits.to_csv("granvik_5k.des", index=False, sep=" ")
```

```python
# now to make the physical parameters file

out_params = pd.DataFrame()

out_params["ObjID"] = out_orbits["ObjID"]
out_params["H_r"] = 15.
out_params["GS"] = 0.15
out_params["u-r"] = 1.
out_params["g-r"] = 0.5
out_params["i-r"] = -0.2
out_params["z-r"] = 0.
out_params["y-r"] = -0.2

    

```

```python
out_params.to_csv("granvik_5k_params.txt", index=False, sep=" ")
```
